```python
prices = []
N = 60*60*24*365
for i in range(N):
    prices.append(100 + i/100)
print(prices[:5])
```

```python
avg = 0.0
for p in prices:
    avg += p/len(prices)
print(avg)
```

```python
import numpy as np
```

```python
prices = 100 + np.arange(N)/100
print(prices[:5])
```

```python
avg = np.mean(prices)
print(avg)
```
